```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
points = np.array([[0.1, 0.6], [0.15, 0.71], [0.08, 0.9], [0.16, 0.85], [0.2, 0.3], [0.25, 0.5], [0.24, 0.1], [0.3, 0.2]])

m1 = points[0]
m2 = points[7]
```

```python
km = KMeans(n_clusters=2, init=np.array([m1,m2]), random_state=42)
```

```python
km.fit(points)
cluster_pred = km.predict(points)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(points[:, 0], points[:, 1], c=cluster_pred)
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], marker='x')
plt.scatter(m1[0],m1[1],c='red')
plt.scatter(m2[0],m2[1],c='red')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('KMeans Clustering')
plt.legend()
plt.show()
```

```python
P6 = np.array([[0.25, 0.5]])
p6_cluster = km.predict(P6)
print(p6_cluster[0]+1)
```
