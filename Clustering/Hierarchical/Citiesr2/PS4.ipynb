```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('cities_r2.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
#now we want to process on the effective literacy rate
data = df[['effective_literacy_rate_total']]
sc = StandardScaler()
scaled_data = sc.fit_transform(data)
```

```python
#now we will plot dendogram 
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10,6))
plt.title('Dendogram')
plt.xlabel('city')
plt.ylabel('distance')
dend = shc.dendrogram(shc.linkage(scaled_data,method='ward'))
```

```python
#now from above dendrogram we can coclude that the clusters are 3
cluster = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
pred = cluster.fit_predict(scaled_data)
```

```python
pred
```

```python
df['Cluster']=pred
df.head()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data,data.index,c=df['Cluster'])
plt.title('Agglomerative Clustering')
plt.xlabel('Cities')
plt.ylabel('Effective Literacy Rate')
plt.show()
```
