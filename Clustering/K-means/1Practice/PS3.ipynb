```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('cities_r2.csv')
df.head()
```

```python
df.shape
```

```python
df.isna().sum().sum()
```

```python
column_to_cluster = ['total_graduates']
selected_data = df[column_to_cluster]
sc = StandardScaler()
scaled_data = sc.fit_transform(selected_data)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['total_graduates'],df.index)
plt.show()
```

```python
k_rng = range(1,25)

sse = []

for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(scaled_data)
    sse.append(km.inertia_)

sse
```

```python
plt.figure(figsize=(10,6))
plt.plot(k_rng,sse,marker='o',color='black',linestyle='dashed')
plt.show()
```

```python
km = KMeans(n_clusters=3,random_state=42)
km.fit(scaled_data)
```

```python
pred = km.predict(scaled_data)
pred
```

```python
df['Cluster'] = pred
df.head()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['total_graduates'],df.index,c=df['Cluster'])
plt.show()
```
