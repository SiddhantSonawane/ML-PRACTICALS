Clustering: K-Means

We have given a collection of 8 points. P1=[0.1,0.6] P2=[0.15,0.71] 
P3=[0.08,0.9] P4=[0.16, 0.85] P5=[0.2,0.3] P6=[0.25,0.5], P7=[0.24,0.1] P8=[0.3,0.2]. 

Perform the k-mean clustering with initial centroids 
as m1=P1 =Cluster#1=C1 and m2=P8=cluster#2=C2. 

Answer the following 
1] Which cluster does P6 belongs to? 
2] What is the population of a cluster around m2? 
3] What is the updated value of m1 and m2?


```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Given points
points = np.array([[0.1, 0.6], [0.15, 0.71], [0.08, 0.9], [0.16, 0.85], [0.2, 0.3], [0.25, 0.5], [0.24, 0.1], [0.3, 0.2]])

# Initial centroids
m1 = points[0]  # P1
m2 = points[7]  # P8
```

```python
# Create KMeans model with 2 clusters and initial centroids
kmeans = KMeans(n_clusters=2, init=np.array([m1, m2]), random_state=42)
```

```python
# Fit the model
kmeans.fit(points)

# Assign clusters to the points
cluster_assignments = kmeans.predict(points)
```

```python
# Visualize the points, clusters, and centroids on a scatter plot
plt.figure(figsize=(8, 6))

# Scatter plot for points
plt.scatter(points[:, 0], points[:, 1], c=cluster_assignments, cmap='viridis', label='Points')

# Scatter plot for centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='X', s=200, c='red', label='Centroids')

# Plot m1 and m2 as initial centroids
plt.scatter(m1[0], m1[1], c='green', marker='o', s=100, label='Initial m1')
plt.scatter(m2[0], m2[1], c='orange', marker='o', s=100, label='Initial m2')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('KMeans Clustering')
plt.legend()
plt.show()
```

```python
# Answers to the questions
# 1] Which cluster does P6 belong to?
P6 = np.array([[0.25, 0.5]])
cluster_P6 = kmeans.predict(P6)
print(f"1] P6 belongs to Cluster C{cluster_P6[0] + 1}")

# 2] What is the population of a cluster around m2?
population_m2 = np.sum(kmeans.labels_ == kmeans.predict([m2]))
print(f"2] The population of the cluster around m2 (C2) is {population_m2}")

# 3] What is the updated value of m1 and m2?
m1_updated = kmeans.cluster_centers_[0]
m2_updated = kmeans.cluster_centers_[1]
print(f"3] Updated value of m1: {m1_updated}")
print(f"   Updated value of m2: {m2_updated}")
```
